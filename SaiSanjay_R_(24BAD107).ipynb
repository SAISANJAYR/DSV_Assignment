```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
plt.rcParams['figure.figsize'] = (8,5)
```

## Load Dataset

```python
df = pd.read_csv('happiness_2017.csv')
df.head()
```

## Dataset Info

```python
df.info()
```

## Descriptive Statistics

```python
df.describe(include='all')
```

## Missing Values

```python
df.isnull().sum()
```

## Histogram

```python
df.select_dtypes(include=np.number).hist(bins=15)
plt.show()
```

## Box Plot

```python
df.select_dtypes(include=np.number).plot(kind='box')
plt.show()
```

## Scatter Plot

```python
plt.scatter(df['Economy..GDP.per.Capita.'], df['Happiness.Score'])
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

## Line Plot

```python
top10 = df.sort_values('Happiness.Score', ascending=False).head(10)
plt.plot(top10['Country'], top10['Happiness.Score'])
plt.xticks(rotation=45)
plt.show()
```

## Bar Plot

```python
plt.bar(top10['Country'], top10['Happiness.Score'])
plt.xticks(rotation=45)
plt.show()
```

## Correlation Heatmap

```python
corr = df.select_dtypes(include=np.number).corr()
plt.imshow(corr)
plt.colorbar()
plt.show()
```
